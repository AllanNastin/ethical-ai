```python
import pickle

with open('triplets.pkl', 'rb') as f:
    triplets = pickle.load(f)
    
# Remove empty lists
triplets = [lst for lst in triplets if lst]

# Keep only the first relation per sentence
triplets = [item[0] for item in triplets]
triplets
```

```python
import json

with open('relations.jsonl', 'w') as file:
    for item in triplets:
        json_str = json.dumps(item)
        file.write(json_str + '\n')
```
