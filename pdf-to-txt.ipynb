```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer
import re


def matches_format(s):
    """Check if a string matches the format of the 'page string' in a pdf"""
    pattern = r'^\d+ \nEN \n$'
    return bool(re.match(pattern, s))


def is_valid(s):
    """Check if a string from the pdf is valid (body) or useless (header/footnotes)"""
    invalid_lines = ['5662/24  \n', 'TREE.2.B \n', 'RB/ek \nLIMITE \n', '  \n', ' \n']
    if s in invalid_lines or matches_format(s):
        return False
    return True


# TODO: Remove footnotes / references from output text
def pdf_to_str(path):
    """Convert a pdf to a string: path=path to pdf"""
    combined_str = ""
    for page_layout in extract_pages(path):
        for element in page_layout:
            if isinstance(element, LTTextContainer):
                # print("--------------------------------------------- NEW TEXT BOX---------------------------------")
                if is_valid(element.get_text()):
                    combined_str += element.get_text()

    return combined_str
    

print(pdf_to_str("ai-act-draft.pdf"))
```
